```python
import pandas as pd

data = {
    "Department": ["IT", "IT", "IT", "IT", "HR", "HR", "HR", "HR"],
    "Salary": [60000, 65000, 70000, 72000, 50000, 52000, 54000, 56000]
}

df = pd.DataFrame(data)
df
```

```python
import matplotlib.pyplot as plt

df.boxplot(column="Salary", by="Department")
plt.title("Salary Distribution by Department")
plt.suptitle("")
plt.ylabel("Salary")
plt.show()
```

```python
from scipy import stats

it_salary = df[df["Department"] == "IT"]["Salary"]
hr_salary = df[df["Department"] == "HR"]["Salary"]

t_stat, p_value = stats.ttest_ind(it_salary, hr_salary)

print("T-statistic:", t_stat)
print("P-value:", p_value)
```

```python
# Decision rule:
# If p-value < 0.05 -> statistically significant difference
```

```python
import numpy as np

means = [it_salary.mean(), hr_salary.mean()]
labels = ["IT", "HR"]

plt.bar(labels, means)
plt.ylabel("Average Salary")
plt.title("Average Salary by Department")
plt.show()
```
