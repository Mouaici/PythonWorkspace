```python
import pandas as pd

data = {
    "Employee": ["Alice", "Bob", "Charlie", "Diana", "Eve"],
    "Age": [25, 30, 35, 28, 32],
    "Salary": [50000, 60000, 70000, 55000, 65000],
    "City": ["London", "Berlin", "Paris", "Madrid", "Rome"],
    "Latitude": [51.5074, 52.5200, 48.8566, 40.4168, 41.9028],
    "Longitude": [-0.1278, 13.4050, 2.3522, -3.7038, 12.4964],
}

df = pd.DataFrame(data)
df
```

```python
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.feature as cfeature

plt.figure(figsize=(10, 6))

ax = plt.axes(projection=ccrs.PlateCarree())

# Add map features
ax.add_feature(cfeature.LAND)
ax.add_feature(cfeature.COASTLINE)
ax.add_feature(cfeature.BORDERS, linestyle=':')

# Plot employees
scatter = ax.scatter(
    df["Longitude"],
    df["Latitude"],
    s=df["Salary"] / 500,        # bubble size = salary
    c=df["Age"],                 # color = age
    cmap="viridis",
    alpha=0.8
)

# Labels
for _, row in df.iterrows():
    ax.text(row["Longitude"] + 0.3, row["Latitude"] + 0.3, row["City"], fontsize=9)

plt.colorbar(scatter, label="Age")
plt.title("HR Analytics: Employee Age & Salary by Office Location")
plt.show()
```
